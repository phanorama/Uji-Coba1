```python
dt['UMUR'] = \
dt['UMUR'].map({'lebih dari 65 tahun':0, 'kurang dari 25 tahun':1, 'antara 25-40 tahun':2, 'antara 40-65 tahun':3})
```

```python
dt['PENGHASILAN'] = \
dt['PENGHASILAN'].map({'lebih dari 3x UMR':0, 'kurang dari UMR':1, 'antara 1-2x UMR':2, 'antara 2-3x UMR':3})
```

```python
dt['KLAIM'] = \
dt['KLAIM'].map({'tidak': 0, 'ya':1})
```

```python
dt['PENGALAMAN_MENGEMUDI'] = \
dt['PENGALAMAN_MENGEMUDI'].map({'kurang dari 10 tahun':0, 'antara 10-20 tahun':1, 'antara 20-30 tahun':2, 'lebih dari 30 tahun':3})
```

```python
dt['PENDIDIKAN'] = \
dt['PENDIDIKAN'].map({'SMA':1, 'tidak ada':0, 'universitas':2})
```

```python
dt['JENIS_KENDARAAN'] = \
dt['JENIS_KENDARAAN'].map({'keluarga':1, 'sport':0})
```
